```python
from pyshacl import validate

# Load the SHACL shape from file
with open('pidinst-sh.ttl', 'r') as f:
    shacl_shape = f.read()

# Load the RDF data graph from file  
with open('non-valid.ttl', 'r') as f:
    data_graph = f.read()

# Validate the data graph against the SHACL shape
conforms, results_graph, results_text = validate(
    data_graph,
    shacl_graph=shacl_shape,
    data_graph_format='turtle',
    shacl_graph_format='turtle',
    inference='none', 
    debug=True,
    serialize_report_graph=True
)

print(conforms)
print(results_text)
```
